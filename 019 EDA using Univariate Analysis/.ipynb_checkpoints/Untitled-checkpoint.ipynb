# Univariate Analysis

- UNi ==> single
- variate ==> variable

- The independent analysis of every column of the data is known as uni-variate analysis
- Generally done through plotting the graphs

```python
import pandas as pd
df = pd.read_csv('train.csv') # titanic 
```

```python
df.head()
```

# 1. Categorical Data

### a. Countplot

```python

import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(df['Survived'])
```

```python
df['Survived'].value_counts().plot(kind = 'bar')
```

## Pie chart

```python
df['Survived'].value_counts().plot(kind = 'pie')
```

```python
df['Pclass'].value_counts().plot(kind = 'pie', autopct = "%.2f") # to get percentage
```

# 2 Categorical data

### a. Histogram
- can make numerica data act as categorical data
- creates bin of a particular range
- Used to generally see data's distribution

```python
plt.hist(df['Age'], bins = 50)
```

### b. DistPlot

```python
sns.distplot(df['Age'])

# thin line is kernal density distribution
# gives PDF ==> probability distribution fn
# plots age vs probability
```

### c. BoxPlot
- provides 5 number summary
![box plot](./BoxPlot.png)
- Tells weather the data is skew or not

```python
sns.boxplot(df['Fare'])
```
